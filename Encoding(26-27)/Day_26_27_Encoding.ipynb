# OneHotEncoding with top categories

```python
counts = cars['brand'].value_counts()
```

from the upper statistcs we make categories till 'Skoda' and include all the categories below it in 'Others' caategory

```python
threshold = 100
```

```python
others = counts[counts<=threshold].index
```

```python
cars['brand'] = cars['brand'].replace(others,'Others')
```

```python
cars['brand'].value_counts()
```

```python
ohe = OneHotEncoder(sparse_output=False)
x = ohe.fit_transform(cars[['brand']])
```

```python
brands = pd.DataFrame(x, columns=ohe.get_feature_names_out())
```

```python
cars = pd.concat([cars.drop('brand', axis=1), brands], axis=1)
```
